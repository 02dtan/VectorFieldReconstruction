```python
import numpy as np
import matplotlib.pyplot as plt
import data.DataGenerator as dg
from data import maps

plt.style.use('figstyle.mplstyle')
```

```python
# Example generation of 1D trajectory with the logistic map, r = 4, scalar initial condition x0

fig, ax = plt.subplots()

X0 = 0.29892
generator = dg.DataGenerator(X0, num_steps = 100, mapname = "logisticmap", map_args = 4)

# Test DataGenerator.generate_series(), plotted below
X = generator.generate_series()

plotx = X[:-1]
ploty = X[1:]

# Test DataGenerator.generate_next(), printed below
print(generator.generate_next(X[-1]))

ax.scatter(plotx, ploty, s=2)
ax.set(xlabel=r'$x_n$', ylabel=r'$x_{n+1}$')
ax.set_title("Logistic map, r = 4")
fig.tight_layout()
```

```python
# Example generation of 2D trajectory with the anosov diffeomorphism, default A = [[2,1],[1,1]], 2D vector initial condition X0

fig, ax = plt.subplots()

X0 = [0.1, 0.3]
generator = dg.DataGenerator(X0, num_steps = 100, mapname = "anosov_diffeo", map_args = None) # anosov_diffeo will take default A = [[2,1],[1,1]]

# Test DataGenerator.generate_series(), plotted below
X = generator.generate_series()

plotx = X[:,0]
ploty = X[:,1]

# Test DataGenerator.generate_next(), printed below
print(generator.generate_next(X[-1]))

ax.scatter(plotx, ploty, s=2)
ax.set(xlabel=r'$x_n$', ylabel=r'$y_n$')
ax.set_title("Anosov diffeomorphism")
fig.tight_layout()
```
